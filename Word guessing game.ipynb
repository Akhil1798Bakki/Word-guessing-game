### The main aim of this project is to build a program which allows user to guess the word
### In this project the user will also provided with Hint for guessing the word
### In this project we have considered 5 words of which the user needs to guess a word
### This project is build in Tkinter platform where the user can interact with the computer 

```python
from tkinter import * # This lines loads tkinter library
import random # this line imports random 

root = Tk() # loads the tkinter
root.title("Guess the character") # Title of the program
root.geometry("500x300") # geometry of the program

# The below class is where the whole Word Guessing game runs
class wordGuessingGame:
    
    read = StringVar() # This is used to manage the values of widget
    attempts = 10 # this specifies the number of attempts
    words = ['Rainbow','Sun', 'People','Tree','Moon'] # these are the words given to the computer 
    answer = random.choice(words)  # a random word will be selected by the computer from this
    
    # The below code is where all the initializations are done for the tkinter window
    def __init__(self):
        
        global read
        
        Label(root, text = "Guess the word!!", font = "Times 20 bold normal ").pack()
        
        self.entry_window = Entry(root, width = 20, borderwidth = 2, font = ("Times", 14)) # This is where the user enters the guessed number
        self.entry_window.pack()
        
        self.btn_check = Button(root, text = "Check", command = self.check_answer)# This is Check button 
        self.btn_check.pack()
        
        self.btn_hint = Button(root, text = "Hint", command = self.give_hint)# This is Hint button 
        self.btn_hint.pack()

        self.btn_quit = Button(root, text = "Quit", command = root.destroy)# This is Quit button, which closes the window
        self.btn_quit.pack()
        
        self.read.set("You have 10 attempts remaining, Good luck")

        self.guess_attempts = Label(root, textvariable = self.read)
        self.guess_attempts.pack()
        
        root.mainloop()
    
    # the below function is used to validate the guessed answer matching the random word selected by the computer
    def check_answer(self):
        
        global attempts
        global answer
        
        self.attempts -= 1
    
        self.guess = self.entry_window.get()# This is where the user entry is stored

        if self.answer == self.guess: # checks the users guess matches with the answer 
            self.read.set("You Won")
            self.btn_check.pack_forget()

        elif self.attempts == 0: # gives the result where there is no attempts remaining
            self.read.set("You have taken all your attempts \n Better Luck next time")
            self.btn_check.pack_forget()

        elif self.guess != self.answer:# Specifies that the user guessed wrong
            self.read.set("Incorrect! you have " + str(self.attempts) + " remaining")
            
    # The below code is used to give Hint to the user
    def give_hint(self):
        
        if self.answer == "Rainbow":
            self.hint = Label(root, text = "HINT \n It is a 7 letter word starts with 'R' and ends with 'W' \n It usually comes after rain")
            self.hint.pack()
            self.btn_hint.pack_forget()

        elif self.answer == "Sun":
            self.hint = Label(root, text = "HINT \n It is a 3 letter word starts with 'S' and ends with 'N' \n Heat comes in mind when we hear this word")
            self.hint.pack()
            self.btn_hint.pack_forget()

        elif self.answer == "People":
            self.hint = Label(root, text = "HINT \n It is a 6 letter word starts with 'P' and ends with 'E' \n This word resembles a Group")
            self.hint.pack()
            self.btn_hint.pack_forget()

        elif self.answer == "Tree":
            self.hint = Label(root, text = "HINT \n It is a 4 letter word starts with 'T' and ends with 'E' \n This word reminds of oxygen")
            self.hint.pack()
            self.btn_hint.pack_forget()

        elif self.answer == "Moon":
            self.hint = Label(root, text = "HINT \n It is a 4 letter word starts with 'M' and ends with 'N' \n It comes every night")
            self.hint.pack()
            self.btn_hint.pack_forget()
            
wordguessinggame = wordGuessingGame()
```
